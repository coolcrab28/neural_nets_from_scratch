# A single neuron




Visually:



```python
inputs = [1.0, 2.0, 3.0, 2.5]
weights = [0.2, 0.8, -0.5, 1.0]
bias = 2.0
output = 0
for i in range(len(inputs)):
    output += inputs[i] * weights[i]
output += bias
```

```python
print(output)
```

# A layer of neurons:




## All neuron-neuron connections have their own weights. All neurons have biases attached to them.


```python
inputs = [1.0, 2.0, 3.0, 2.5]
weights = [[0.2, 0.8, -0.5, 1.0], [0.5, -0.91, 0.26, -0.5], [-0.26, -0.27, 0.17, 0.87]]
biases = [2.0, 3.0, 0.5]
outputs = [0] * len(weights)
```

```python
for connection in range(len(weights)):
    for i in range(len(inputs)):
        outputs[connection] += inputs[i] * weights[connection][i]
    outputs[connection] += biases[connection]
print(outputs)
```

```python
x = zip(weights, biases)
print(list(x))
```
